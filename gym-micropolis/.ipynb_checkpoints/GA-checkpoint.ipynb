```python
import numpy as np
import random
from gym_micropolis.envs.corecontrol import MicropolisControl
from gym_micropolis.envs.tilemap import TileMap,zoneFromInt
from gi.repository import Gtk as gtk
from time import sleep
```

# Genetic Algorithm to build micropolis city

## Define Constants 

```python
# Size of the map, in reality it can be a bit bigger because we allow to build a 4*4 block on the last case 
MAP_H = 10
MAP_W = 10
GEN_SIZE = 10 # Size of the Genome - Number of buildings 
N_POPULATION = 500
N_STEPS_EVALUATION = 100
```

## Building list 

```python
# We only take a few buildings at first
buildings = ['NuclearPowerPlant','Residential','Commercial','Industrial','Road']
```

## Generate Random Genoms

```python
genomes = [[[np.random.randint(1,MAP_H),np.random.randint(1,MAP_H),random.choice(buildings)] for i in range(GEN_SIZE)] for p in range(N_POPULATION)]
```

## Cross 2 Genomes

```python
def cross_genomes(g1,g2) : 
    """
    Take 2 genomes and build a new one using crossing
    """
    r = np.random.randint(len(g1))
    return(g1[0:r] + g2[r:])
```

## Mutate a genome

```python
def mutate_genome(g1) :
    """
     Take a genome and apply a random transformation
    """
    p = np.random.randint(len(g1))
    g1[p] = [np.random.randint(1,MAP_H),np.random.randint(1,MAP_W),random.choice(buildings)]
```

# Evaluate a Genome

```python
def build_city(g1,display) : 
    """
    Build a city for a given genome an return it 
    """
    m = MicropolisControl(MAP_H,MAP_W,display=display)
    m.clearMap()
    # Build
    for b in g1 : 
        m.doTool(b[0],b[1],b[2])
    return m
```

```python
def evaluate(g1,display=False) : 
    """
        Build and evaluate a city
    """
    m = build_city(g1,display)
    for i in range(N_STEPS_EVALUATION) :
        m.engine.simTick()
        m.render()
    if display :
        m.win1.destroy()
    pop = m.engine.cityPop
    m.close()
    return pop
```

```python
%%capture
populations = []
for g in genomes :
    populations.append(evaluate(g,display=False))
```
